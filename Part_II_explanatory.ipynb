## I. Investigation Overview

In this analysis, we explore the 'Loan Data from Prosper' dataset to uncover insights about loan details and borrower profiles. Our goal is to understand key aspects such as loan amounts, interest rates, and credit grades. We aim to address the following key questions:

## II. Dataset Overview

The 'Loan Data from Prosper' dataset provides detailed records of loans, including information about loan amounts, borrower rates, loan statuses, and borrower profiles. The dataset includes the following key attributes:

- <strong>Loan Amount</strong>: The amount of the loan as stated in the dataset.
- <strong>Borrower APR</strong>: The Annual Percentage Rate for the loan.
- <strong>Credit Grade</strong>: The credit rating assigned to the borrower.
- <strong>Listing Category</strong>: The category of the loan listing, such as Debt Consolidation, Home Improvement, etc.

The data allows us to analyze loan characteristics and borrower information to identify trends and patterns.

## III. Visualizations

### A. Distribution of Loan Amounts

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
df = pd.read_csv('prosperLoanData.csv')

# Drop rows where 'LoanOriginalAmount' is NaN
df = df.dropna(subset=['LoanOriginalAmount'])

# Convert 'LoanOriginationDate' to datetime
df['LoanOriginationDate'] = pd.to_datetime(df['LoanOriginationDate'])

# Extract year and month from 'LoanOriginationDate'
df['OriginationYear'] = df['LoanOriginationDate'].dt.year
df['OriginationMonth'] = df['LoanOriginationDate'].dt.month

# Plot distribution of loan amounts
plt.figure(figsize=(10, 6))
sns.histplot(df['LoanOriginalAmount'], bins=50, kde=True)
plt.title('Distribution of Loan Amounts')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
```

<strong>Observation:</strong>

The histogram shows that most loans fall within a certain range of amounts, with a few loans having significantly higher values. The distribution may be right-skewed, with fewer high-value loans.

### B. Number of Loans by Credit Grade

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='CreditGrade')
plt.title('Number of Loans by Credit Grade')
plt.xlabel('Credit Grade')
plt.ylabel('Count')
plt.show()
```

<strong>Observation:</strong>

The count plot indicates the distribution of loans across different credit grades. Higher credit grades are likely to have fewer loans compared to lower credit grades.

### C. Loan Amount vs. Borrower APR

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BorrowerAPR', y='LoanOriginalAmount', data=df)
plt.title('Loan Amount vs. Borrower APR')
plt.xlabel('Borrower APR')
plt.ylabel('Loan Amount')
plt.show()
```

<strong>Observation:</strong>

There may be a correlation between loan amount and borrower APR, with varying patterns across different APR values.

### D. Loan Amount by Listing Category

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='ListingCategory (numeric)', y='LoanOriginalAmount')
plt.xticks(rotation=90)
plt.title('Loan Amount by Listing Category')
plt.xlabel('Listing Category')
plt.ylabel('Loan Amount')
plt.show()
```

<strong>Observation:</strong>

Loan amounts vary significantly across different listing categories, with some categories showing higher loan amounts on average.

### E. Hourly Loan Amount Patterns

```python
plt.figure(figsize=(14, 8))
sns.histplot(data=df, x='OriginationMonth', hue='CreditGrade', multiple='stack', kde=True)
plt.title('Monthly Loan Amount Patterns by Credit Grade')
plt.xlabel('Month of Origination')
plt.ylabel('Count')
plt.show()
```

<strong>Observation:</strong>

The facet plot reveals how loan amounts and credit grades vary across different months, providing insights into seasonal trends and credit grade distribution.
